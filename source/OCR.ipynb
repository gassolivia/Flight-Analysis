```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Function to create a new PDF with text from images
def ocr_pdf(input_pdf_path, output_pdf_path):
    # Open the original PDF
    pdf_document = fitz.open(input_pdf_path)
    new_pdf = fitz.open()  # Create a new PDF

    # Iterate through each page
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        # Create a new page in the new PDF
        new_page = new_pdf.new_page(width=page.rect.width, height=page.rect.height)
        
        # Extract images from the page
        images = page.get_images(full=True)
        text = ""

        for img_index, img in enumerate(images):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image = Image.open(io.BytesIO(image_bytes))
            
            # Perform OCR on the image
            text += pytesseract.image_to_string(image)

        # Insert the recognized text into the new PDF
        new_page.insert_text((50, 50), text, fontsize=12)

    # Save the new PDF with text
    new_pdf.save(output_pdf_path)
    new_pdf.close()
    pdf_document.close()

# Example usage
ocr_pdf('../hidden/input.pdf', '../hidden/output_with_text.pdf')
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Function to create a new PDF with text from images
def ocr_pdf(input_pdf_path, output_pdf_path):
    # Open the original PDF
    pdf_document = fitz.open(input_pdf_path)
    new_pdf = fitz.open()  # Create a new PDF

    # Iterate through each page
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        # Create a new page in the new PDF
        new_page = new_pdf.new_page(width=page.rect.width, height=page.rect.height)
        
        # Extract images from the page
        images = page.get_images(full=True)
        text = ""

        if not images:
            print(f"No images found on page {page_num + 1}.")
            continue  # Skip to the next page if no images are found

        for img_index, img in enumerate(images):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image = Image.open(io.BytesIO(image_bytes))
            
            # Perform OCR on the image
            recognized_text = pytesseract.image_to_string(image)
            print(f"Recognized text from image {img_index + 1} on page {page_num + 1}: {recognized_text}")

            text += recognized_text + "\n"  # Add a newline for separation

        # Check if any text was recognized
        if text.strip():
            # Insert the recognized text into the new PDF
            new_page.insert_text((50, 50), text, fontsize=12)
        else:
            print(f"No text recognized on page {page_num + 1}.")

    # Save the new PDF with text
    new_pdf.save(output_pdf_path)
    new_pdf.close()
    pdf_document.close()

# Example usage
ocr_pdf('../hidden/input.pdf', '../hidden/output_with_text.pdf')
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Function to convert PDF pages to images and perform OCR
def pdf_to_text_via_ocr(input_pdf_path, output_pdf_path):
    # Open the original PDF
    pdf_document = fitz.open(input_pdf_path)
    new_pdf = fitz.open()  # Create a new PDF

    # Iterate through each page
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        
        # Convert the page to a pixmap (image)
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

        # Perform OCR on the image
        recognized_text = pytesseract.image_to_string(img)
        print(f"Recognized text from page {page_num + 1}: {recognized_text}")

        # Create a new page in the new PDF
        new_page = new_pdf.new_page(width=page.rect.width, height=page.rect.height)

        # Insert the recognized text into the new PDF
        if recognized_text.strip():
            new_page.insert_text((50, 50), recognized_text, fontsize=12)
        else:
            print(f"No text recognized on page {page_num + 1}.")

    # Save the new PDF with text
    new_pdf.save(output_pdf_path)
    new_pdf.close()
    pdf_document.close()

# Example usage
pdf_to_text_via_ocr('../hidden/input.pdf', '../hidden/output_with_text.pdf')
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image, ImageFilter
import io

# Specify the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'  # Update this path as necessary

# Function to convert PDF pages to images and perform OCR
def pdf_to_text_via_ocr(input_pdf_path, output_pdf_path):
    # Open the original PDF
    pdf_document = fitz.open(input_pdf_path)
    new_pdf = fitz.open()  # Create a new PDF

    # Iterate through each page
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        
        # Convert the page to a pixmap (image) with increased resolution
        pix = page.get_pixmap(matrix=fitz.Matrix(2, 2))  # Increase DPI
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

        # Preprocess the image
        img = img.convert('L')  # Convert to grayscale
        img = img.filter(ImageFilter.MedianFilter(size=3))  # Apply a median filter

        # Perform OCR on the image
        recognized_text = pytesseract.image_to_string(img, config='--psm 6')  # Use appropriate PSM
        print(f"Recognized text from page {page_num + 1}: {recognized_text}")

        # Create a new page in the new PDF
        new_page = new_pdf.new_page(width=page.rect.width, height=page.rect.height)

        # Insert the recognized text into the new PDF
        if recognized_text.strip():
            new_page.insert_text((50, 50), recognized_text, fontsize=12)
        else:
            print(f"No text recognized on page {page_num + 1}.")

    # Save the new PDF with text
    new_pdf.save(output_pdf_path)
    new_pdf.close()
    pdf_document.close()

# Example usage
pdf_to_text_via_ocr('../hidden/input.pdf', '../hidden/output_with_text.pdf')
```
